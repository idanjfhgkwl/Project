```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 시각화 사전 준비
sns.set_context("talk")
sns.set_style("white")

# 한글 사용 준비
plt.rcParams['font.family']='NanumGothic'
plt.rcParams['axes.unicode_minus'] = False
```

```python
fig, ax = plt.subplots(figsize=(20, 9))

# 지원자 수 
ax = df_sn[["seoul", "gyeonggi"]].plot.bar(ax=ax)

# 응시자 수
#ax.plot(df_sn["QLTWTR_INSPCT_DE"], df_sn["num_test"], color=color_test, label="응시자 수")

# 범례 추가
ax.legend()
```

```python
fig, ax = plt.subplots(figsize=(20, 9))
ax.plot(df_sn.loc[121:123, "datatime"], df_sn.loc[121:123,"seoul"],'o-', label = 'seoul')
ax.plot(df_sn.loc[121:123, "datatime"], df_sn.loc[121:123,"gyeonggi"],'o:', label = 'gyeonggi')
ax.plot(df_sn.loc[121:123, "datatime"], df_sn.loc[121:123,"incheon"],'o--', label = 'incheon')


plt.axhline(y=10, color='r', linewidth=3)

# 주변 테두리를 제거
ax.spines["left"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.spines["top"].set_visible(False)

ax.grid(axis="y")



ax.legend()
```
