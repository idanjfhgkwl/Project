```python
fig = plt.figure(figsize=(12,7))

widths = [4, 4]
heights = [1, 4]

# widths 와 행,열 수가 맞아야한다

### 1. gridspec preparation
spec = fig.add_gridspec(ncols=2, nrows=2, width_ratios=widths, height_ratios=heights)

### 2. setting axes
axs = {}
for i in range(len(heights)*len(widths)):
    axs[i] = fig.add_subplot(spec[i//len(widths), i%len(widths)])
    axs[i].text(0.5, 0.5, f"axs[{i}]", fontdict={"horizontalalignment":"center", "color":"gray"})
```

```python
fig = plt.figure(figsize=(12,7))

widths = [4, 4]
heights = [1, 4]

# 1. gridspec preparation
spec = fig.add_gridspec(ncols=2, nrows=2, width_ratios=widths, height_ratios=heights)

# 2. setting axes
axs = {}
for i in range(len(heights)*len(widths)):
    axs[i] = fig.add_subplot(spec[i//len(widths), i%len(widths)])
    
# 4. scatterplot - seoul vs gyeonggi
sns.scatterplot("서울","경기", data = df_sn,hue ="측정항목", ax = axs[2])

# histogram
sns.kdeplot(y="서울", data = df_sn, hue = "측정항목", ax = axs[0],legend=False)
axs[0].set_xlim(axs[3].get_xlim())
axs[0].set_xlabel('')
axs[0].set_xticklabels([])
axs[0].spines["left"].set_visible(False)
axs[0].spines["top"].set_visible(False)
axs[0].spines["right"].set_visible(False)

# scatterplot - seoul vs incheon
sns.scatterplot("서울","인천", data = df_sn, hue="측정항목", ax = axs[3])

sns.kdeplot("인천", data=df_sn, hue="측정항목", ax=axs[1], legend=False)
axs[1].set_xlim(axs[3].get_xlim())
axs[1].set_xlabel('')
axs[1].set_xticklabels([])
axs[1].spines["left"].set_visible(False)
axs[1].spines["top"].set_visible(False)
axs[1].spines["right"].set_visible(False)
```
