```python
%matplotlib inline

# 시각화 사전 준비
sns.set_context("talk")
sns.set_style("white")

# 한글 사용 준비
plt.rcParams['font.family']='NanumGothic'
plt.rcParams['axes.unicode_minus'] = False
```

```python
fig, ax = plt.subplots(figsize=(20, 9))
 
ax = df[["서울", "경기","인천"]].plot.bar(ax=ax)


plt.xticks(rotation=0)

#plt.xticks(range(0,max(ax),5))    # x축의 눈금의 범위 : 눈금의 범위 + 눈금의 간격
#plt.plot(x,y)

# 주변 테두리를 제거
ax.spines["left"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.spines["top"].set_visible(False)

ax.grid(axis="y")



title_font = {
    'fontsize': 16,
    'fontweight': 'bold'
}
plt.title('경기도권 24시간 기준 3일 측정', fontdict=title_font, loc='center', pad=40)

for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x() + p.get_width()/2.,
            height+ 0.5, int(height), ha='center', va = 'bottom',size=15)

plt.ylabel('㎍/m³',fontsize = 23)



plt.text(1.15, 38, '2020-06-22',fontweight = "bold")
plt.text(3.2, 38, '2020-06-23',fontweight = "bold")
plt.text(5, 38, '2020-06-24',fontweight = "bold")
plt.axvline(x=1.5, ymin=0.005, ymax=0.7, color = 'r')
plt.axvline(x=3.5, ymin=0.005, ymax=0.7, color = 'r')
plt.axvline(x=5.4, ymin=0.005, ymax=0.7, color = 'r')


# 범례 추가
ax.legend()
plt.show()
```
