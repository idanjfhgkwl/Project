## 데이터 불러오기

```python
df_air_jun6 = pd.read_csv("data/대기오염시도별실시간평균정보_2020년_05월.csv")
df_air_jun6.head()
```

## 열이름 한글로 바꾸기

```python
df_air_jun6.columns = ["기간","측정항목","데이터 구분","서울","부산","대구","인천","광주","울산","대전","경기","강원","충북","충남",
                       "전북","전남","경북","경남","제주","세종"]
df_air_jun6.head()
```

```python
df_air_jun6.info()
```

## 기간 순으로 정렬하기

```python
df_air_jun6 = df_air_jun6.sort_values(['기간'])
df_air_jun6
```

## 필요없는 열 삭제하기 (기간, 측정항목, 지역만 남기기)

```python
df_air_jun6 = df_air_jun6.drop(["데이터 구분","부산","대구","광주","울산","대전","강원","충북","충남","전북","전남","경북","경남","제주","세종"], axis = 1)
df_air_jun6
```

```python
df_air_jun6.columns
```

## 기간이 6월 22일 ~ 24일, 측정항목이 PM10, PM2.5인 데이터만 남기기

```python
df_air_jun6_date = df_air_jun6[((df_air_jun6['기간'] >= '2020-06-22') & (df_air_jun6['기간'] <= '2020-06-24'))
                              & ((df_air_jun6['측정항목'] == 'PM10') | (df_air_jun6['측정항목'] == 'PM2.5'))]
df_air_jun6_date
```

## pandas melt 함수를 이용해서 데이터 재구조화하기

출처: [Python] 데이터 재구조화(reshape) : pd.melt()  
https://rfriend.tistory.com/278

```python
df_air_jun6_district = pd.melt(df_air_jun6_date, id_vars=['기간','측정항목'],
                               var_name='지역', value_name='농도')
df_air_jun6_district = df_air_jun6_district[['지역','측정항목','기간','농도']]
df_air_jun6_district
```

## 지역을 서울-경기-인천 순으로 어떻게 정렬하지? (진행중)

## 측정항목별 데이터프레임 나누기

```python
df_pm10 = df_air_jun6_district[df_air_jun6_district['측정항목'] == 'PM10']
df_pm2_5 = df_air_jun6_district[df_air_jun6_district['측정항목'] == 'PM2.5']
df_pm2_5
```

## 시각화

```python
# 시각화 사전 준비
sns.set_context("talk")
sns.set_style("white")

# 한글 사용 준비
plt.rcParams['font.family']='NanumGothic'
plt.rcParams['axes.unicode_minus'] = False
```

```python
# subplots 3개 만들기 (y축 공유 = 세로 동일)
fig, axes = plt.subplots(ncols=2, figsize=(20, 10), sharey=True)
plt.suptitle("'6월' 경기도권 24시간 기준 3일 측정",fontsize=30)

# 농도별(PM10, PM2.5) 그래프 나누기
df_pm10_lineplot=sns.lineplot(x="기간", y="농도", hue="지역",  palette="Set2",
                data=df_pm10, ax=axes[0], marker='o');
df_pm2_5_lineplot=sns.lineplot(x="기간", y="농도", hue="지역", palette="Set2",
                data=df_pm2_5, ax=axes[1], marker='o');

df_pm10_lineplot.grid(axis="y", color='gainsboro')
df_pm2_5_lineplot.grid(axis="y", color='gainsboro')

# x축 제목 변경
axes[0].set_xlabel("PM10", fontsize=22, fontweight = 'bold')
axes[1].set_xlabel("PM2.5", fontsize=22, fontweight = 'bold')

# y축 이름 변경
axes[0].set_ylabel("㎍/m³")
axes[1].set_ylabel("")

# y축 단위 설정
import matplotlib.ticker as ticker
axes[0].yaxis.set_major_locator(ticker.MultipleLocator(5))
axes[0].set_ylim(14,52)

# 그래프가 들어있는 각 박스 삭제
sns.despine(bottom=False, left=False, ax=axes[0])
sns.despine(bottom=False, left=False, ax=axes[1])

df_pm10_lineplot.spines["left"].set_visible(False)
df_pm2_5_lineplot.spines["left"].set_visible(False)

df_pm10_lineplot.spines["bottom"].set_visible(False)
df_pm2_5_lineplot.spines["bottom"].set_visible(False)

# 1번째 범례 삭제 (hide legend)
axes[0].legend([],[], frameon=False)

# 그래프 위 수치 나타내기
for p in axes[0].patches:
    axes[0].annotate("%.f" % p.get_height(),
      (p.get_x()+p.get_width()/2, p.get_height() + 1), ha='center', va='top', color='dimgrey', size=18)

for p in axes[1].patches:
    axes[1].annotate("%.f" % p.get_height(),
      (p.get_x()+p.get_width()/2, p.get_height() + 1), ha='center', va='top', color='dimgrey', size=18)
```
